```python
import os
import subprocess

import pandas as pd
import numpy as np
import graphviz
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn import tree
```

```python
#citesc setul de date
datasets = pd.read_csv('ObesityDataSet.csv')
```

```python
#afisez primele 5 
datasets.head()
```

```python
datasets.dtypes
```

```python
le_ObesityLevel = LabelEncoder()
le_ObesityLevel.fit(datasets['NObeyesdad']) 
data_encoded_ObL = le_ObesityLevel.transform(datasets['NObeyesdad'])
data_encoded_ObL
```

```python
le_ObesityLevel.classes_
```

```python
datasets_1 = datasets.apply(le_ObesityLevel.fit_transform) # for all the attributes
datasets_1
```

```python
X = datasets_1[datasets.columns.drop('NObeyesdad')]
Y = datasets_1['NObeyesdad']
```

```python
X
```

```python
Y
```

```python
#Impart dataset-ul in Training set si Test set
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.25)
```

```python
X_Train
```

```python
Y_Train
```

```python
print(len(X_Train))
print(len(X_Test))
print(len(Y_Train))
print(len(Y_Test))
```

```python
#Feature Scaling
stdScaler = StandardScaler() #Standardize features by removing the mean and scaling to unit variance
X_Train = stdScaler.fit_transform(X_Train)
X_Test = stdScaler.transform(X_Test)
```

```python
X_Train
```

```python
Y_Train
```

```python
dt1 = tree.DecisionTreeClassifier(criterion="entropy")
dt1
```

```python
dt1.fit(X_Train, Y_Train)
```

```python
pred = dt1.predict(X_Test)
pred
```

```python
dt1.predict_proba(X_Test)
```

```python
#METRICA 1: Matricea de confuzie
#METRICA 2: Accuracy;
#METRICA 3: Precision
#METRICA 4: Recall
#METRICA 5: F1
from sklearn import metrics
matrix = confusion_matrix(Y_Test, pred)
print("Confusion Matrix:")
print(matrix)
print("Accuracy score:",metrics.accuracy_score(Y_Test, pred))
print("Precision score:",metrics.precision_score(Y_Test, pred))
print("Recall score:",metrics.recall_score(Y_Test, pred))
print("F1 score:",f1_score(Y_Test, pred))
```
